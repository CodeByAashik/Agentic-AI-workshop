```python
from langchain_ollama import ChatOllama
```

```python
model = ChatOllama(
    model="llama3.2",
    temperature=0.7,
    num_predict=300,   # STOP after 300 tokens
)
```

```python
from langchain_core.prompts import PromptTemplate

TEMPLATE = """Write about tourism potential of {country}"""
prompt = PromptTemplate(
    template=TEMPLATE,
    input_variables=["country"]
)
```

```python
formatted_prompt = prompt.format(
    country="Italy"
)

formatted_prompt
```

```python
response = model.invoke(formatted_prompt)
response
```
